## Dataset:
Function ***read_data*** is used to read the data from the file located at the path provided. Here, I have used the ***vehicles.csv*** dataset.

```python
X, y, class_map, _, _ = read_data('../../datasets/vehicles.csv')
```

```python
X_train, X_test, y_train, y_test = split_data(X, y)
```

## Experiment: Logistic Regression vs SVM vs KNN
The below function call converts the provided data into binary data and performs classification using 3 classifiers:
- Logistic Regression
- Support Vector Machine
- K Nearest Neighbour

Following the classification, a plot is generated that shows the percentage of samples predicted with each unique probability (of predicted class) comparing the 3 classifiers.

```python
compare(models=['logistic', 'svm', 'knn'], X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test)
```

### Observations:
The plot generated above gives us an insight how good a classifier is in terms of how confident it's predictions are. For example, we see that the SVM classifier performs poorly and the histogram concurs with this observation as most if not all of the classifier's predictions have a confidence score between 0.5 and 0.6 only. On the other hand, we see that both KNN and Logistic regression classifiers perform much better, and this is again complemented by the indications of high proportions of the samples being predicted with a probability/confidence of 0.9 - 1.0.
