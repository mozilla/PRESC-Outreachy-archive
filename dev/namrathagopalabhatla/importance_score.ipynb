```python
from classification_models import Classifier
from data_processing import read_data, split_data
from importance_score import importance_score

import warnings
%matplotlib inline
```

```python
warnings.filterwarnings('ignore')
```

## Dataset:
Function ***read_data*** is used to read the data from the file located at the path provided. Here, I have used the ***vehicles.csv*** dataset.

```python
X, y, class_map, _, _ = read_data('../../datasets/vehicles.csv')
```

# Datapoint Importance Score
The below experiment tries out different combinations of parameters. They are listed as follows:
- **75-25** Data Split
    - **Accuracy** as metric
    - **F1 Score** as metric
- **60-40** Data Split
    - **Accuracy** as metric
    - **F1 Score** as metric
- **80-20** Data Split
    - **Accuracy** as metric
    - **F1 Score** as metric

```python
model = Classifier(model='logistic')
```

## Experiment with 75-25 Split

```python
X_train, X_test, y_train, y_test = split_data(X, y)
```

### Accuracy as metric

```python
importance_score(model=model, 
                 X_train=X_train, 
                 y_train=y_train, 
                 X_test=X_test,
                 y_test=y_test,
                 metric='accuracy'
                )
```

### F1 Score as metric

```python
importance_score(model=model, 
                 X_train=X_train, 
                 y_train=y_train, 
                 X_test=X_test,
                 y_test=y_test,
                 metric='score'
                )
```

## Experiment with 60-40 Split

```python
X_train, X_test, y_train, y_test = split_data(X, y, 0.6)
```

### Accuracy as metric

```python
importance_score(model=model, 
                 X_train=X_train, 
                 y_train=y_train, 
                 X_test=X_test,
                 y_test=y_test,
                 metric='accuracy'
                )
```

### F1 Score as metric

```python
importance_score(model=model, 
                 X_train=X_train, 
                 y_train=y_train, 
                 X_test=X_test,
                 y_test=y_test,
                 metric='score'
                )
```

## Experiment with 80-20 Split

```python
X_train, X_test, y_train, y_test = split_data(X, y, 0.8)
```
